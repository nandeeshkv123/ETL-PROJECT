## Transform

Using exchange_rate and the `exchange_rates.csv` file find the exchange rate of USD to GBP. Write a transform function that

1. Changes the `Market Cap (US$ Billion)` column from USD to GBP
2. Rounds the Market Cap (US$ Billion)` column to 3 decimal places
3. Rename `Market Cap (US$ Billion)` to `Market Cap (GBP$ Billion)`


```python
import pandas as pd

def transform(exchange_rate):
    # Load the CSV file into a DataFrame
    df = pd.read_csv('exchange_rates.csv')
    
    # Find the exchange rate of USD to GBP
    usd_to_gbp_rate = exchange_rate.loc['USD', 'GBP']
    
    # Convert USD to GBP in the 'Market Cap (US$ Billion)' column
    df['Market Cap (US$ Billion)'] = df['Market Cap (US$ Billion)'] * usd_to_gbp_rate
    
    # Round the 'Market Cap (US$ Billion)' column to 3 decimal places
    df['Market Cap (US$ Billion)'] = df['Market Cap (US$ Billion)'].round(3)
    
    # Rename the 'Market Cap (US$ Billion)' column to 'Market Cap (GBP$ Billion)'
    df.rename(columns={'Market Cap (US$ Billion)': 'Market Cap (GBP$ Billion)'}, inplace=True)
    
    return df
```
